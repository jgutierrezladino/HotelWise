- Creamos funsiones para hacer un analisis de sentimiento en el comentario de los usuarios.

```python
def map_score_to_1_5(score):
    # Mapear el puntaje de sentimiento a la escala de 1 a 5
    mapped_score = ((score + 1) / 2) * 4 + 1
    return mapped_score

def agregar_score_sentimiento(df):
    # Inicializar el analizador de sentimientos VADER
    sia = SentimentIntensityAnalyzer()

    # Calcular el puntaje de sentimiento para cada comentario y agregarlo como una nueva columna
    df['score_sentimiento'] = df['text'].apply(lambda comentario: map_score_to_1_5(sia.polarity_scores(comentario)['compound']) if comentario is not None else 0.0)

    return df
```

- Accionamos la funsion de analisis de sentimiento.

```python
df_h_analisis = agregar_score_sentimiento(df_h)
```
